```python
X = df.drop(columns=['target'])#independent variable: contains allcolums except target column
y = df['target']#dependnet or target value: contains only target column
print(X.shape)
print(y.shape)
```

```python
# x_train,x_test,y_train,y_test <------ order is important
# random test ----> to pick data randomly to split data into test and train
x_train,x_test,y_train,y_test = train_test_split(X,y,random_state=0, test_size=0.3) #30% will be kept for test
print(x_train.shape)
print(x_test.shape)
```

```python
clf = tree.DecisionTreeClassifier()  #configuring the model
clf.fit(x_train,y_train) #Training the model
y_train_pred = clf.predict(x_train) #predicting the tained data
y_test_pred = clf.predict(x_test)
```

```python
# helper function
def plot_confusionmatrix(y_train_pred,y_train,dom):
    print(f'{dom} Confusion matrix')
    cf = confusion_matrix(y_train_pred,y_train)
    sns.heatmap(cf,annot=True,cmap='Blues', fmt='g')#For g and G , the maximum number of significant digits
    plt.tight_layout()
    plt.show()
```

```python
print(f'Train score {accuracy_score(y_train_pred,y_train)}')
print(f'Test score {accuracy_score(y_test_pred,y_test)}')
```

```python
plot_confusionmatrix(y_train_pred,y_train,dom='Train')
plot_confusionmatrix(y_test_pred,y_test,dom='Test')
```
